# 正規分布

```python
# 数値計算に使うライブラリ
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats

# グラフを描画するライブラリ
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()

# グラフをjupyter Notebook内に表示させるための指定
%matplotlib inline
```

```python
print(stats.norm.pdf(loc = 5.84, scale = 0.25, x = 5.6))
print(stats.norm.pdf(loc = 5.84, scale = 0.25, x = 5.0))
```

```python
x_plot = np.arange(start = 4.8, stop = 6.8, step = 0.01)
#確率密度
plt.plot(
    x_plot, 
    stats.norm.pdf(x = x_plot, loc = 5.84, scale = 0.25),
)
#累積
plt.plot(
    x_plot, 
    stats.norm.cdf(x = x_plot, loc = 5.84, scale = 0.25),
)
```

# t分布

```python
x_plot = np.arange(start = -6.0, stop = 6.0, step = 0.01)
#確率密度
plt.plot(
    x_plot, 
    stats.t.pdf(x = x_plot, df=5),
)
```

```python
y1 = stats.t.pdf(x=x_plot, df=1)
y3 = stats.t.pdf(x=x_plot, df=3)
y10 = stats.t.pdf(x=x_plot, df=10)
norm = stats.norm.pdf(x=x_plot,loc=0,scale=1)

plt.plot(x_plot,y1, label='ν=1')
plt.plot(x_plot,y3, label='ν=3')
plt.plot(x_plot,y10, label='ν=10')
plt.plot(x_plot,norm, label='norm')
plt.ylim(0,0.45); plt.xlim(-5,5)
plt.legend()
```
